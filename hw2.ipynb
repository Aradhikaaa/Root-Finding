## Defining the function and the derivative of it
### $f(x)= x^3 - 7x^2 + 14x -5$
### $f'(x) = 3x^2 - 14x +14$
The derivative is needed for the Newton Raphson Method

```python
# Defining the function

def func(x):
    return (x**3 - 7* x**2 + 14*x -5)

def dfunc(x):
    return (3*x**2 - 14*x + 14)
```

## Bisection Method

```python
m = methods()
n=1

x1=0
x2=1
print(f"The intervals are {{{x1}, {x2}}}")
tol=1e-8
n1,r1 = m.bisection(func,x1,x2,tol)
if r1 is not None:
    print("The root is: ", r1)
    print("The number of iterations is: ", n1)
```

### Sensitivity of iteration count to initial guess

```python
print("BISECTION: Effect of initial interval on iteration count")
print('\n')

intervals = [(0, 1), (0, 0.5), (-5, 5), (0.4, 0.45)]

for (x1, x2) in intervals:
    print(f"For Interval [{x1},{x2}]")
    n1, r1 = m.bisection(func, x1, x2, tol)
    if r1 is not None:
        print("  Interval [",x1,",", x2,"] : root = ",r1,", iterations = ",n1)


print("NEWTON-RAPHSON: Effect of initial guess on iteration count")
print('\n')
```

## Problematic initial guesses for Newton-Raphson

The derivative of the function is $f'(x) = 3x^2 - 14x + 14$.

Setting $f'(x) = 0$ gives two critical points: 

$x = \frac{14 + \sqrt{28}}{6}$ and $x = \frac{14 - \sqrt{28}}{6}$.

These correspond to extremum points of the function:

- $x = \frac{14 + \sqrt{28}}{6}$ → **local minimum**, because $f''(x) = 6x - 14 > 0$ here  
- $x = \frac{14 - \sqrt{28}}{6}$ → **local maximum**, because $f''(x) = 6x - 14 < 0$ here  

Starting Newton-Raphson near these points can cause problems because the derivative is very small or zero, leading to very large steps or division by zero.

```python
# Critical points where derivative = 0
crit1 = (14 + np.sqrt(28)) / 6   # local minimum
crit2 = (14 - np.sqrt(28)) / 6   # local maximum

print("Critical points of f(x):")
print("  Local minimum at x =", crit1)
print("  Local maximum at x =", crit2)

# Newton-Raphson with tricky initial guesses
x0 = crit1
print("Initial guess x0 =", x0, "(at local minimum)")
n, root = m.newton(func, dfunc, x0, tol)
print("  Root =", root, ", iterations =", n)

x0 = crit1 + 0.001
print("Initial guess x0 =", x0, "(just past local minimum)")
n, root = m.newton(func, dfunc, x0, tol)
print("  Root =", root, ", iterations =", n)

x0 = crit2
print("Initial guess x0 =", x0, "(at local maximum)")
n, root = m.newton(func, dfunc, x0, tol)
print("  Root =", root, ", iterations =", n)

x0 = crit2 - 0.001
print("Initial guess x0 =", x0, "(just before local maximum)")
n, root = m.newton(func, dfunc, x0, tol)
print("  Root =", root, ", iterations =", n)
```
